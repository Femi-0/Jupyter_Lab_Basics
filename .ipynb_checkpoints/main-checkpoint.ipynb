```python
#import imperative to make pandas and numpy_finanacial libraries available
import pandas as pd
import numpy_financial as np
#location of budget csv file
path = "./budget_data.csv"
```

```python
#initiate csv as a variable
df= pd.read_csv(path)
```

```python
# displays stored csv
df
```

```python
#Check no repeated month in list
df['Date'].duplicated().sum()

#Determine Total Month
total_months = len(df['Date'])
total = df['Profit/Losses'].sum()

#Create changes column
df['Changes'] = ''
difference_list = []
difference_list.insert(0,0)
def get_difference_column(df_in):
    i = 0
    while i < (len(df['Date'])-1):
        current_difference = (df['Profit/Losses'][i+1]) - (df['Profit/Losses'][i])
        difference_list.append(current_difference)
        i += 1

get_difference_column(df)
df['Changes'] = difference_list

#Get average of changes
avg = "{:.2f}".format((df['Changes'].sum())/(len(df['Changes'])-1))

#Get greatest increase/decrease in profits
inc = df['Changes'].max()
dec = df['Changes'].min()

#Get index of months
inc_month = df[df['Changes']==inc].index.values
dec_month = df[df['Changes']==dec].index.values

#Get months
increase_month = list(df['Date'][inc_month])
decrease_month = list(df['Date'][dec_month])

print(f"Total Months: {total_months}")
print(f"Total: ${total}")
print(f"Average Change: ${avg}")
print(f"Greatest increase in profits: {increase_month[0]} ${inc}")
print(f"Greatest decrease in earnings: {decrease_month[0]} ${dec}")

#Retrun data to csv file
df.to_csv('budget_data_with_changes.csv', index=False)
```
